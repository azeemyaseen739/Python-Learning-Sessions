### 1. Lists
##### **What are Lists?**

**Fixed Vs Dynamic Size:**

### 4. Acessing Items from a List

```python
# Indexing
L = [1,2,3,4,5]
print(L[0]) # positive indexing, index number start from 0 and move from left to right
print(L[-1]) # negative indexing, index number start from -1 and move from right to left
# if we have have the 2D list then, to fetch a number we use double square brackets
L = [1,2,3,[4,5]]
print(L[3][0]) # through positive indexing
print(L[-1][-2]) # through negtative indexing
# if we have 3D list then, to fect a number we use triple sqaure brackets
L = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(L[0][0][0]) # positive indexing
print(L[-1][-2][-2]) # negative indexing
```

```python
# slicing, extracting multiple items from a list
L = [1,2,3,4,5]
# positive slicing
print(L[0:3])
print(L[3:])
print(L[:4])
print(L[:])
print(L[0::2])
print(L[0:4:2]) # 2 is the step size
# negative slicing
print(L[::-1]) # to reverse the list
print(L[-5:])
print(L[-1:0:-1])
```

### 10. List Comprehension
 - List Comprehension provides a concise way of creating lists.
 - newlist = [expression **for** item **in** iterable **if** condition **==** True]\
**Advantages of List Comprehension:**
 - More time-efficient and space-efficient than loops.
 - Require fewer lines of code.
 - Transform iterative statement into a formula.

```python
# Add 1 to 10 numbers to a list
L = []
for i in range(1,11):
    L.append(i)
print(L)
# in this code snippet we use 4 lines of code, but with list comprehension we do this in 1 line
L = [i for i in range(1,11)]
print(L)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = 3

L = []
for i in v:
    L.append(i*s)
print(L)
# this is the 4 line code snippet, but we do this in 1 line
print([i*s for i in v])
```

```python
# Add squares
v = [1,2,3,4,5]

L = []
for i in v:
    L.append(i*i)
print(L)
# this is the 4 line code snipet but you do this in 1 line
print([i*i for i in v])
```

```python
# print all numbers divisible by 5 in the range of 1 to 50
print([i for i in range(1,50+1) if i%5 == 0]) 
# This is the power of list comprehension, you generate the whole program in 1 line
```

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']
print([i for i in languages if i[0] == 'p'])
```

```python
# Nested if with list comprehension
# add new list from my_fruits and items if the fruits exist in basket and also starts with 'a'
basket = ['apple','guava','cherry','banana']
my_fruit = ['apple','kiwi','grapes','banana']

my_fruits = [i for i in basket if i in my_fruit and i[0] == 'a']
print(my_fruits)
```

```python
# print a (3,3) matric using list comprehension -> neted list comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian products -> list comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]

L = []
for i in L1:
    for j in L2:
        L.append(i*j)
print(L)

[i*j for i in L1 for j in L2]
```

### 12. Zip:
 - The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item is each passed iterator are paired together.
 - If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
# write a program to add items of 2 lists indexwise
L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

list(zip(L1,L2))
print([i+j for i,j in zip(L1,L2)])
```

```python
# Create 2 lists from a given list where 
# 1st list will contain all the odd numbers from the original list and
# the 2nd one will contain all the even numbers
L = [1,2,3,4,5,6]

odd_list = []
even_list = []

for i in L:
    if i % 2 == 0:
        even_list.append(i)
    else:
        odd_list.append(i)

print('the even numbers list',even_list)
print('the odd numbers list',odd_list)
```
